```python
import pandas as pd
import re
from collections import Counter
```

```python
text = """
data analytics is very important
data science is interesting
machine learning is powerful
data analytics helps businesses
machine learning improves predictions
analytics and machine learning are related
"""
```

```python
text = text.lower()
words = re.findall(r'\b\w+\b', text)
```

```python
word_freq = Counter(words)
word_freq
```

```python
def autocomplete(prefix):
    suggestions = [word for word in word_freq if word.startswith(prefix)]
    return suggestions

autocomplete("data")
```

```python
def autocorrect(word):
    if word in word_freq:
        return word
    else:
        return "No suggestion found"

autocorrect("analytcs")
```

Conclusion:
This project demonstrates basic autocomplete and autocorrect functionality using text data. Word frequency analysis helps predict and suggest words efficiently.
